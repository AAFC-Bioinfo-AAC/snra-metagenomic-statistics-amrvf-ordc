```python
# Filtering MEGARes and VFDB results for the MAGS
# Author: Dr. Haley Sanderson haley.sanderson@agr.gc.ca
# Copyright: Government of Canada
# License: MIT
# Version 0.1
import os
import pandas as pd
```

```python
from pathlib import Path
#create an empty dataframe
all_megares_results = []
#Determine the path that will lead to your mobsuite results files for all HLF genomes
megares_results_directory = "./"
for megares_output in Path(megares_results_directory).glob('metabat_bins.*_megares_output'):

    bin_name = megares_output.parts[0]

    print(f"Parsing {megares_output}")
#Read tables in megares output files with new header we designate
    bin_megares_results = pd.read_table(megares_output, 
                                       sep='\t', 
                                       header=0, 
                                       names=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore", "qseq", "sseq"], 
                                            usecols=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore", "qseq", "sseq"])
#Put Plasmid_name (the name of the isolate) into a column in the table
    bin_megares_results['Bin_Pathname'] = bin_name
#Append the table to your empty dataframe
    all_megares_results.append(bin_megares_results)
#concatentate all the tables together
all_megares_results = pd.concat(all_megares_results)
#Print to see new table
print(all_megares_results)
```

```python
all_megares_results.head()
```

```python
nosnp = all_megares_results[ all_megares_results[ "sseqid" ].str.contains( "RequiresSNPConfirmation" )==False ]
```

```python
nosnp
```

```python
megares_nosnp_80 = nosnp[nosnp['pident'] > 80]
```

```python
megares_nosnp_80
```

```python
crosstab_megares = pd.crosstab(megares_nosnp_80['Bin_Pathname'], megares_nosnp_80['sseqid'])
crosstab_megares
```

```python
megares_nosnp_80 = megares_nosnp_80.set_index("Bin_Pathname")
```

```python
megares_nosnp_80.to_csv("MAG_blast_megares_results_nosnp_80.csv")
crosstab_megares.to_csv("MAG_blast_megares_nosnp_80_crosstab.csv")
```

```python
from pathlib import Path
#create an empty dataframe
all_vfdb_results = []
#Determine the path that will lead to your mobsuite results files for all HLF genomes
vfdb_results_directory = "./"
for vfdb_output in Path(megares_results_directory).glob('metabat_bins.*_vfdb_output'):

    bin_name = vfdb_output.parts[0]

    print(f"Parsing {vfdb_output}")
#Read tables in megares output files with new header we designate
    bin_vfdb_results = pd.read_table(vfdb_output, 
                                       sep='\t', 
                                       header=0, 
                                       names=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore", "qseq", "sseq"], 
                                            usecols=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore", "qseq", "sseq"])
#Put Plasmid_name (the name of the isolate) into a column in the table
    bin_vfdb_results['Bin_Pathname'] = bin_name
#Append the table to your empty dataframe
    all_vfdb_results.append(bin_vfdb_results)
#concatentate all the tables together
all_vfdb_results = pd.concat(all_vfdb_results)
#Print to see new table
print(all_vfdb_results)
```
