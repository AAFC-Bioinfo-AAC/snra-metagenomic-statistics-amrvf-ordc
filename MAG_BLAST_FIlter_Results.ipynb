```python
from pathlib import Path
#create an empty dataframe
all_nt_results = []
#Determine the path that will lead to your mobsuite results files for all HLF genomes
nt_results_directory = "./"
for nt_output in Path(nt_results_directory).glob('metabat_bins.*_nt_db_output'):

    mag_name = nt_output.parts[0]

    print(f"Parsing {nt_output}")
#Read tables in mobsuite output files with new header we designate
    mag_nt_results = pd.read_table(nt_output, 
                                       sep='\t', 
                                       header=0, 
                                       names=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore", "qseq", "sseq"], 
                                            usecols=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore", "qseq", "sseq"])
#Put mag_name (the name of the isolate) into a column in the table
    mag_nt_results['MAG_Pathname'] = mag_name
#Append the table to your empty dataframe
    all_nt_results.append(mag_nt_results)
#concatentate all the tables together
all_nt_results = pd.concat(all_nt_results)
#Print to see new table
print(all_nt_results)
```

```python
nt_80_id = all_nt_results[all_nt_results['pident'] > 90]
nt_80_id
```

```python
nt_80_evalue = nt_80_id[nt_80_id['evalue'] < 1e-5]
nt_80_evalue
```

```python
nt_80_evalue.to_csv("Unidentified_MAG_BLAST_NT_90_1e-5.csv")
```
